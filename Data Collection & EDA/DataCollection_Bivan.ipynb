```python
# Importing billboard to load data from Billboard.com
import billboard
# Importing pandas to convert data into DataFrame
import pandas as pd 
```

```python
#artists: top 10 artists in all years.
artists = [] 
#year: year wise loading songs.
year = 2021 
for y in range(2022-2006):
    #chart: It contains data from Billboard which include artists names and title of songs
    chart = billboard.ChartData('Top-Artists', year=year)
    cnt = 0
    for chr in chart:
        cnt += 1
        if chr.artist not in artists:
                artists.append(chr.artist)
        #cnt: Stop itration when we get top 10 artists every year.        
        if cnt == 10:
            break
    
    year -= 1
#Collected Top 80 artists from 2006 - 2021
print(artists)
len(artists)
```

```python
def getData(type, data, artists, titles, year=None, date=None, label=None): #type means type of chart we get from Billboard.
    if date==None: #data is a list of dictionary we collocet data into.
        HotChart = billboard.ChartData(type, year=year) #year to use it when collect popular data.
    else:
        HotChart = billboard.ChartData(type, date=date) #date to use it when collect unpopular data.
    for hot in HotChart:
        # To collect the name of artists that exists in the artist's list and choose another song for the same artist.
        # This is to avoid repeat songs and enter artist doesn't exist in artists list.
        if hot.title not in titles and hot.artist in artists: #titles is a list of songs name use it to avoid repeat songs.
                titles.append(hot.title)
                data.append({'Artist': hot.artist, 'Title': hot.title, 'Label': label}) #label is 1 if popular and 0 if unpopular.
    
    return data, titles
```

```python
data = []
titles = []
year = 2021
for i in range(2022-2006):
    data, titles = getData('Hot-100-Songs', data, artists, titles, year, label=1) # popular songs
    year -= 1
```

```python
#We collect popular data from 2006 to 2021
print(len(data))
print(len(artists))
print(len(titles))
```

```python
# Splited the year into 4 splits to get a lot of songs with the same artists which I choose and used the same techniques for every year.
dates = ['2021-12-01','2021-08-01','2021-04-01','2021-01-01',
        '2020-12-01','2020-08-01','2020-04-01','2020-01-01',
        '2019-12-01','2019-08-01','2019-04-01','2019-01-01',
        '2018-12-01','2018-08-01','2018-04-01','2018-01-01',
        '2017-12-01','2017-08-01','2017-04-01','2017-01-01',
        '2016-12-01','2016-08-01','2016-04-01','2016-01-01',
        '2015-12-01','2015-08-01','2015-04-01','2015-01-01',
        '2014-12-01','2014-08-01','2014-04-01','2014-01-01',
        '2013-12-01','2013-08-01','2013-04-01','2013-01-01',
        '2012-12-01','2012-08-01','2012-04-01','2012-01-01',
        '2011-12-01','2011-08-01','2011-04-01','2011-01-01',
        '2010-12-01','2010-08-01','2010-04-01','2010-01-01',
        '2009-12-01','2009-08-01','2009-04-01','2009-01-01',
        '2008-12-01','2008-08-01','2008-04-01','2008-01-01',
        '2007-12-01','2007-08-01','2007-04-01','2007-01-01',
        '2006-12-01','2006-08-01','2006-04-01','2006-01-01']
for date in dates:
    data, titles = getData('hot-100', data, artists, titles, date=date, label=0) # To get unpopular songs
```

```python
#Printed unpopular data from 2006 to 2021
print(len(data))
print(len(artists))
print(len(titles))
```
